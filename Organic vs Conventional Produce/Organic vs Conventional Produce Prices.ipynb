```python
fdf13 = pd.read_csv("fruit2013.csv")
fdf13.columns = fdf13.loc[0]
fdf13 = fdf13.drop(fdf13.index[0])
fdf13
```

```python
fdf12 = pd.read_csv("fruit2012.csv", keep_default_na=False)
fdf12.columns = fdf12.loc[0]
fdf12 = fdf12.drop(fdf12.index[0])
fdf12
```

```python
cfdf12 = fdf12[fdf12['Organic/ Conventional'] == 'Conv']
ofdf12 = fdf12[fdf12['Organic/ Conventional'] == 'Org']
cfdf13 = fdf13[fdf13['Organic/ Conventional'] == 'Conv']
ofdf13 = fdf13[fdf13['Organic/ Conventional'] == 'Org']
cvdf12 = vdf12[vdf12['Organic/ Conventional'] == 'Conv']
ovdf12 = vdf12[vdf12['Organic/ Conventional'] == 'Org']
cvdf13 = vdf13[vdf13['Organic/ Conventional'] == 'Conv']
ovdf13 = vdf13[vdf13['Organic/ Conventional'] == 'Org']
```

```python
def month(x, y):
    z = 0
    u = 0
    for i in range(len(x)):
        if x.iloc[i][y] != 'N\\A':
            z += float(x.iloc[i][y])
            u += 1
    avg = z/u
    return avg

month(cfdf12, 'Jan-12')
```

```python
month(ofdf12, 'Jan-12')
```

```python
d = []
for columns in cfdf12.columns[5:17]:
    c = {'2012 Conventional Fruit': month(cfdf12, columns), '2012 Organic Fruit': month(ofdf12, columns) }
    d.append(c)
fruit2012 = pd.DataFrame(d)
```

```python
fruit2012
```

```python
ps = fruit2012.plot()
ps.set_xlabel("Month (Jan = 0)")
ps.set_ylabel("Price (in $)")
```
